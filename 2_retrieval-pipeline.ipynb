```python
from langchain_chroma import Chroma

# from langchain_openai import OpenAIEmbeddings
from langchain_huggingface import HuggingFaceEmbeddings
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
# Define the persistent directory for Chroma database
persistent_directory = "db/chroma_db"

# Load embeddings and vector store
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Initialize Chroma vector store with the specified persistent directory and embedding model
db = Chroma(
    persist_directory=persistent_directory,
    embedding_function=embedding_model,
    collection_metadata={"hnsw:space": "cosine"},
)
```

```python
# Search for relevant documents
query = "How much did Microsoft pay to acquire GitHub?"

# Create a retriever from the Chroma vector store with a specified number of results to return (k=5)
retriever = db.as_retriever(search_kwargs={"k": 5})

# retriever = db.as_retriever(
#     search_type="similarity_score_threshold",
#     search_kwargs={
#         "k": 5,
#         "score_threshold": 0.3  # Only return chunks with cosine similarity ≥ 0.3
#     }
# )

# Retrieve relevant documents based on the query
relevant_docs = retriever.invoke(query)
```

```python
print(f"User Query: {query}")
# Display results
print("--- Context ---")
for i, doc in enumerate(relevant_docs, start=1):
    print(f"Document {i}:\n{doc.page_content}\n")
```
